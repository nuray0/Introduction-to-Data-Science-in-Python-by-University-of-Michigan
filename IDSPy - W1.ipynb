Lambda is a way of creating anonymous functions in Python - they are the same as other functions, but they have no name

There is only one expression to be evaluated in lambda

Simple, no name, single line

```
my_func = lambda a, b, c: a*b
```

```
my_func(1,2,3)
```

The return of the lambda is a function reference

You can't have default values for lambda parameters 

You can't have complex logic inside the lambda itself (because you're limited to single expression)

```
people = ['Dr. Christopher Brooks', 'Dr. Kevyn Collins-Thompson', 'Dr. VG Vinod Vydiswaran', 'Dr. Daniel Romero']

def split_title_and_name(person):
    return person.split()[0] + ' ' + person.split()[-1]

#option 1
for person in people:
    print(split_title_and_name(person) == (lambda person: person.split()[0] + ' ' + person.split()[-1])(person))
print()
#option 2
list(map(split_title_and_name, people)) == list(map(lambda person: person.split()[0] + ' ' + person.split()[-1], people))
```

### Manipulating Text with Regular Expressions (regexes) - written in condensed formatting language

Reg Ex is a pattern given to regex processor with some source data -> parses source data (using that pattern), and returns chunks of text back to the data scientist for further manipulation.

? Why we need it?

- to check wether a patter exists within the source data.

-  to clean the source data using a pattern generally through string splitting

- to get all instances of a complex pattern from some source data

```
import re
```

```
text = 'this is a good day'

if re.search('good', text): # Looks for a match anywhere in the string
  print('found!')
else:
  print('not found(')
```

```
text = 'this is a good day'

if re.match('th', text): # checks for a match
  print('matches!')
else:
  print('does not match(')

if re.match('g', text): # checks for a match
  print('matches!')
else:
  print('does not match(')
```

### Groups

```
for item in re.finditer("([\w ]*)(\[edit\])", text):
  print(item.groups())
```

```
for item in re.finditer("([\w ]*)(\[edit\])", text):
  print(item.group(1))
```

```
for item in re.finditer("([\w ]*)(\[edit\])", text):
  print(item.group(0))
```

group(0) - whole match

group(n) - group we are interested in

```
for item in re.finditer("([\w ]*)(\[edit\])", text):
  print(item.group(2))
```

```
for item in re.finditer("([\w ]*)(\[edit\])", text):
  print(item.group(3))
```

labeling the groups and looking as them as dictionary

```
(?P<name>)
# ( - starts the group
# ?P - indicates extension to basic regexes
# <name> - dictionary key
# (\w\d)+ plus here requires at least one of each w and d
```

```
for item in re.finditer("(?P<title>[\w ]*)(?P<edit_link>\[edit\])", text):
  print(item.groupdict()['title'])
```

. for any single character thats not newline
\d for any digit
\s for any whitespace character, like spaces and tabs

### Look-ahead and Look-behind matching

```
import re
# If you want to capture the text befor eedit but do not care about edit itself
# you can use look ahead with ?= syntax
for item in re.finditer("(?P<title>[\w ]+)(?=\[edit\])", text):
  print(item)
```
